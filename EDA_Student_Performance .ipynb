```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
Student_Data = pd.read_csv('StudentsPerformance.csv')
```

```python
Student_Data
```

```python
Student_Data.sort_values(by = 'math score', ascending = False )
```

```python
Student_Data.isnull().sum()
```

```python
Student_Data.info()
```

```python
Student_Data.describe()
```

From the table above we can see that math score mean is lower than reading and writing score, but it doesn't have much difference

```python
Student_Data.corr()
```

- There is a high positive correlation between writng and reading score indicating an equal ability for each subject, if student have a high score in reading there is a high probability of having a high score in writing 
- Math score doesnt have high correlation with another subject which mean if someone have high score in math doesn't mean they have high score in another subject

```python
sns.countplot(x='gender', data = Student_Data)
```

```python
sns.pairplot(Student_Data,hue='gender')
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.boxplot(x='gender',y='math score',data=Student_Data)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.boxplot(x='gender',y='reading score',data=Student_Data)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.boxplot(x='gender',y='writing score',data=Student_Data)
plt.title('WRITING SCORES')
plt.show()
```

- Female have a high score in reading and writing compare to male
- Male have a high score in math compare to female

```python
plt.figure(figsize=(10,7))
sns.set_style('whitegrid')
sns.barplot(x = "parental level of education", y = "reading score", hue = "gender", data = Student_Data)
plt.xticks(rotation=45)
plt.show()
```

Based on the bar plot between parental level of education and scores in each subject, students who have parent with higher education tend to have higher scores

- In Reading score, female who do not completed a course still have similiar performance with male who done the course
- In Writng score, female who do not completed a course still have similiar performance with male who done the course
